```python
#  All Standlaone Base Models
XBase = dataset.iloc[:,0:9].values
X = XBase

X = pd.DataFrame(X)
y = dataset.iloc[:,9].values

X = pd.concat([X, pd.get_dummies(dataset['season'], prefix='Season', drop_first = True)], axis=1)
X = pd.concat([X, pd.get_dummies(dataset['day'], prefix='Day', drop_first = True)], axis=1)
X = pd.concat([X, pd.get_dummies(dataset['time'], prefix='Time', drop_first = True)], axis=1)
X = X.drop(X.columns[[0,1,4]], axis = 1)


#Splitting the data into Training Set and Test Set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)

#Normalizing the features
from sklearn.preprocessing import StandardScaler
sc_X1 = StandardScaler()
X_train = sc_X1.fit_transform(X_train)
X_test = sc_X1.transform(X_test)

#Model #1 :  LinearRegression
from sklearn.linear_model import LinearRegression
lrObject = LinearRegression()
lrObject.fit(X_train, y_train)

#Model #2 :  Decision Tree Object
from sklearn.tree import DecisionTreeRegressor
DTObj = DecisionTreeRegressor()
DTObj.fit(X_train,y_train) 

#Model #3 : RandomForest Object
from sklearn.ensemble import RandomForestRegressor
RFObj = RandomForestRegressor(n_estimators=10)
RFObj.fit(X_train,y_train)


#Model #4 : Ridge Regression

from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
ridge = Ridge()
parameters = {'alpha': [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20]}
RDObj = GridSearchCV(ridge, parameters, cv=10)
RDObj.fit(X_train, y_train)

#Model #5 : Lasso Regression
from sklearn.linear_model import Lasso
lasso = Lasso()
parameters = {'alpha': [1e-15, 1e-10, 1e-8, 1e-4, 1e-3,1e-2, 1, 5, 10, 20]}
LSObj = GridSearchCV(lasso, parameters,  cv = 10)
LSObj.fit(X_train, y_train)


#Prediction on the Test Set
y_pred_LR = lrObject.predict(X_test)
y_pred_DT = DTObj.predict(X_test)
y_pred_RF = RFObj.predict(X_test)
y_pred_RD = RDObj.predict(X_test)
y_pred_LS = LSObj.predict(X_test)


print(" All Models - Root Mean Squarred Error")

from sklearn.metrics import mean_squared_error
from math import sqrt

rmse_LR = sqrt(mean_squared_error(y_test, y_pred_LR))
rmse_DT = sqrt(mean_squared_error(y_test, y_pred_DT))
rmse_RF = sqrt(mean_squared_error(y_test, y_pred_RF))
rmse_RD = sqrt(mean_squared_error(y_test, y_pred_RD))
rmse_LS = sqrt(mean_squared_error(y_test, y_pred_LS))

print("Standalone - All features: RMSE - Linear Regression          ", rmse_LR)
print("Standalone - All features: RMSE - Decision Tree              ", rmse_DT)
print("Standalone - All features: RMSE - Random Forest              ", rmse_RF)
print("Standalone - All features: RMSE - Ridge                      ", rmse_RD)
print("Standalone - All features: RMSE - Lasso                      ", rmse_LS)
```
